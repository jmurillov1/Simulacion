```python
from collections import Counter
import json
res=[]
suma=0
for i in range(6):
    for j in range(6):
        suma=(i+1)+(j+1)
        res.append(suma)
c=Counter(res)
for i in sorted(c.items()):
    print("Valor: ", i[0] , " Porcentaje: ",round((i[1]/11)*100,2), " %")
```

```python
from collections import Counter
import json
import random
import matplotlib.pyplot as plt
resultado=[]
def calcular(num):
    for i in range(num):
        dado1= random.randint(1, 6)
        dado2= random.randint(1, 6)
        resultado.append(dado1+dado2)
    val=Counter(resultado)
    print(json.dumps(val,indent=2,sort_keys=True))
    plt.title('Histograma para '+ str(num) +' valores')
    plt.hist(resultado,bins=num, width=0.5, alpha=0.7, edgecolor = 'black')
    plt.grid(axis='y', alpha=0.75)
    plt.xlabel('Valor',fontsize=15)
    plt.ylabel('Frequencia',fontsize=15)
    plt.xticks(range(0, max(val.keys())+1))
    plt.show()
    for i in sorted(val.items()):
        print("Valor: ", i[0] , " Porcentaje: ",round((i[1]/num)*100,2), " %")
```

```python
calcular(100)
```

```python
calcular(1000)
```
